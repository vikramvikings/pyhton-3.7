# Time comparison

```python
import time
import numpy as np
start =time.time()
l=[x*x for x in range(10)]
print(l)
end = time.time()
print('time with lists:',end-start)
start =time.time()
a=np.arange(10000)
a=a*a
end = time.time()
print('time with np arrays:',end-start)
```

# Stacking Arrays

```python
a = np.arange(6)
b=np.arange(6)
np.hstack([a,b])
```

```python
np.vstack([a,b])
```

```python
a = np.arange(6).reshape(2,3)
b=np.arange(6).reshape(2,3)
np.hstack([a,b])
```

```python
a = np.arange(6).reshape(2,3)
b=np.arange(3)
np.vstack([a,b])
```
