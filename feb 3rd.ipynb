```python
np.random.seed(10)
a=np.random.randint(10,200,50)
a
```

create new array<br>
a>50 and <150 - good<br>
a<50 - bad<br>
a>150 - excellent

```python
def number_cat(n):   
    if n<50:
        return 'bad'
    elif n>=50 and n<=150:
        return 'good'
    else:
        return 'excellent'

```

```python
 number_cat(50)   
```

```python
number_cat_vect= np.vectorize(number_cat)
```

```python
number_cat_vect(a)
```

```python
import pandas as pd
cat = pd.Series(number_cat_vect(a))
cat
```

```python
cat.value_counts()
```

```python
y_values=cat.value_counts()
x_values=cat.value_counts().index
plt.bar(x_values,y_values)
```

```python
p1=plt.bar(x_values,y_values,color='gry')
plt.legend(p1)
```

```python
plt.bar(x_values,y_values,color='gry')
colors={'good':'green','red':'bad','excellent':'yellow'}
labels=list(colors.keys())
handels=[plt.Rectangle((0,0),1,1,color=colors[label])
            for label in labels]
plt.legend(handels,labels)
```
