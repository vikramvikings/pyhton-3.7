# regular expression

```python
import re
s='vikram'
re.search('k',s)
```

```python
print(re.search('k','hello'))
```

```python
re.match('k','kl')
```

```python
re.findall('k',s)
```

```python
names=['ram','raven','rayes','griffin','bellamy']
#find character 'e'

for name in names:
    if re.search('e',name):
        print(name)
    
```

```python
#find the names with starts with 'ra' and atleast 4 characters

for name in names:
    if re.match('ra..',name):
        print(name)
```

```python
#fimd the name whose length is greater than 4 or equal

for name in names:
    if re.search('....',name):
        print(name)
```

```python
print(re.search('[abc]','apple'))
print(re.search('[abc]','bob'))
print(re.search('[abc]','bcb'))
print(re.search('[abc]','nac'))
```

```python
'b,r,g'
```

```python
for name in names:
    if re.search('[brg]',name):
        print(name)
```

# {min  max}

```python
print(re.search('.{2,4}','he'))
print(re.search('.{2,4}','vicky'))
print(re.search('.{2,4}','v'))
```

# \d digit

```python
print(re.search('\d','marks 20'))
print(re.search('\d{2}','marks 20'))
print(re.search('\d{3}','marks 20'))
print(re.search('\d','marks'))
```

# \D

```python
print(re.search('\D','345'))
print(re.search('\D','123onethree'))
```

# \s single space

```python
print(re.search('\s','first name'))
print(re.search('\s','     first name'))
```

# \w[a-z,A-Z,0-9]

```python
print(re.search('\w','var'))
print(re.search('\w','345'))
print(re.search('\w','$45'))
print(re.search('\w','ba'))
print(re.search('\w','_5'))
```
