# Text preprocessing : stemming, stop-words removal and lemmatization

- begin by removing html tags
- remove any punctuations or set of any special characters
- check if word is made up of english letters and is not alpha numeric
- convert the word to lower case
- remove stop words
- fiinally snowball stemming the word

```python
# find sentenances containing html tags
import re
i=0
for sent in data['message'].values:
    if(len(re.findall('<.*?>',sent))):
        print(i)
        print(sent)
        #break;
    i+=1
```

```python
import nltk
nltk.download('stopwords')
import re as stdre
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer
stop= set(stopwords.words('english'))   #set of stopwords
print(stop)
```

```python
sno=nltk.stem.SnowballStemmer('english')
print(sno.stem('thankful'))
```

```python
def cleanhtml(sentence):
    cleanr=stdre.compile('<.*?>')
    cleantext=stdre.sub(cleanr,' ',sentence)
    return cleantext
```

```python
cleanhtml(data['message'][2297])
```

```python
def cleanpunc(sentence):
    cleaned = stdre.sub('[?|!|\'|"|#]',r'',sentence)
    cleaned=stdre.sub('[.|,|)|(|\|/]',r' ' ,cleaned)
    return cleaned
```

```python
data=pd.read_csv("SMSSpamCollection",'\t',names=['spam','message'])
i=0
str1=' '
final_string=[]
for sent in data["message"].values:
    filtered_sentence=[]
    sent=cleanhtml(sent) # remove the html tags
    for w in sent.split():
        for cleaned_words in cleanpunc(w).split():
            #print(cleaned_words)
            if((cleaned_words.isalpha()) & (len(cleaned_words)>2)):
                #print(cleaned_words)
                if(cleaned_words.lower() not in stop):
                    #print(cleaned_words)
                    s=(sno.stem(cleaned_words.lower())).encode('utf-8')
                    #print(s)
                    filtered_sentence.append(s)
                else:
                    continue
            else:
                continue
    # print(filtered_sentence)\n",
    str1= b" ".join(filtered_sentence)
    final_string.append(str1)
    i+=1
data['message']=final_string    
#final_string
```
